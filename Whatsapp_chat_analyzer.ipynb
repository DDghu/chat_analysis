# Here we extract out whatsapp text file and using data preprocessing convert the datas into pandas data and other things we analusis it

```python
import re #for regular expression
import pandas as pd
```

```python
f=open('Whatsapp_chat1','r',encoding='utf-8') #using file handling in python and make sure to take utf-8 orelse not work
```

```python
data=f.read() #read all the datas in the form of string and store in data
```

```python
print(data)
```

```python
pattern='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s' #make regular expression pattern using he above message for splitting 
```

```python
messages=re.split(pattern, data)[1:]
messages
```

```python
dates=re.findall(pattern,data)
dates
```

```python

# Original list of date and time strings
datetime_strings = dates

# Function to convert a single date and time string to the desired format
def convert_datetime_string(datetime_string):
    # Step 1: Remove extra spaces and the trailing dash
    datetime_string = datetime_string.strip().rstrip('-')
    
    # Step 2: Convert to datetime object using pd.to_datetime
    datetime_obj = pd.to_datetime(datetime_string, format='%m/%d/%y, %H:%M ')
    print(datetime_obj)
    # Step 3: Format the datetime object to the desired output format
    formatted_datetime = datetime_obj.strftime('%Y-%m-%d %H:%M:%S')
    return formatted_datetime

# Use list comprehension to apply the conversion function to all date and time strings
converted_datetimes = [convert_datetime_string(dt_str) for dt_str in datetime_strings]

converted_datetimes



```

```python
df=pd.DataFrame({'user_message': messages, 'message_date': converted_datetimes})
# convert message_date typez
df['message_date']=pd.to_datetime(df['message_date'], format='%Y-%m-%d %H:%M:%S')

df.rename(columns={'message_date': 'date'}, inplace=True)
df.head()
```

```python
df.shape
```

```python
#separates users and messages
users=[]
messages=[]
for message in df['user_message']:
    entry=re.split('([\w\W]+?):\s', message)
    if entry[1:]:#user name
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
df['User']=users
df['message']=messages
df.drop(columns=['user_message'],inplace=True)
df.head(30)
        
```

```python
#Now we split and store all the parts of the date 
df['year']=df['date'].dt.year # .dt._is the atribute
df['month']=df['date'].dt.month_name()
df['day']=df['date'].dt.day
df['hour']=df['date'].dt.hour
df['minute']=df['date'].dt.minute
```

```python
# we make a new column as period and we change hour like 16 to 6-17 ,23 to 23-00
period=[]
for hour in df[['day_name','hour']]['hour']:
    if hour == 23:
        period.append(str(hour)+"-"+str('00'))
    elif hour == 0:
        period.append(str('00')+"-"+str(hour+1))
    else:
        period.append(str(hour)+"-"+str(hour+1))
```

```python
df['period']=period
```
